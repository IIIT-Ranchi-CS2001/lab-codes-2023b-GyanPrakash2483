1. You are managing a student club and need to keep track of members and their activities.  
The user will input the details of each member, including their name, age, and the activity they participate in (e.g., Music, Sports, Art).  
You need to:  
a. Create a set of unique activities.  
b. Count how many students are enrolled in each activity.  
c. Display the names of students grouped by their activity.  
d. Allow the user to enter member details until they decide to stop.  

```python
records = []

while True:
    name = input("Name: ")
    if("END" == name):
        break
    age = int(input("Age: "))
    activity = input("Activity: ")
    if("END" == activity):
        break
    records.append(dict({
        "name": name,
        "age": age,
        "activity": activity
    }))

#print(records)

activities = set()
for record in records:
    activities.add(record.get("activity"))

#print(activities)

for activity in activities:
    names = [record.get("name") for record in records if record.get("activity") == activity]
    print(activity + "(" + str(len(names)) + ")" + ": " + str(names))
```

7. Write a program to construct a user defined function sort_list() to sort the strings present in the list based on their length.

```python
def sort_list(strlist: list) -> list:
    for i in range(0, len(strlist)):
        for j in range(i, len(strlist)):
            if(len(strlist[j]) < len(strlist[i])):
                strlist[i], strlist[j] = strlist[j], strlist[i]
    return strlist

print(sort_list(["hi", "hello", "bye"]))
```

9. Create a module with name my_module to incorporate the following functions.
(minimum use of built-in functions)  
a. My_fact(): To find the factorial of all numbers that are passed in the function call. Function shall return a list of integers. (Number of elements that are passed must be of variable length)   
b. My_psq(): To find the perfect square numbers that are passed in the function call. The function shall return a list of integers. (Number of elements that are passed must be of variable length)   
c. Call both the functions from the main program and try the function call for list, set or tuple as its argument

```python
# filename my_module.py (ignored because jupyter notebook, using class instead)

import math

class my_module:
    def My_fact(iterable) -> list:
        def factorial(n: int) -> int:
            if(n == 0):
                return 1
            else:
                return n * factorial(n - 1)
        return [factorial(n) for n in iterable]

    def My_psq(iterable) -> list:
        def isPerfectSquare(n: int) -> int:
            if(math.sqrt(n) == math.floor(math.sqrt(n))):
                return True
            return False            
        return [n for n in iterable if isPerfectSquare(n)]   

print(my_module.My_fact([1, 2, 3, 4, 5]))
print(my_module.My_fact({1, 2, 3, 4, 5})) 
print(my_module.My_fact((1, 2, 3, 4, 5))) 

print(my_module.My_psq([1, 2, 3, 4, 5]))
print(my_module.My_psq({1, 2, 3, 4, 5})) 
print(my_module.My_psq((1, 2, 3, 4, 5))) 

```
