Q1 Show party wise seat share for following results of the Assembly Elections 2023 in  
 - Two different pie charts on two different plots. Party with highest percentage should be shown as slightly detached ( show the percentage seat share on each wedge )
 - Two pie charts as subplots on the same figure object
 - As a bar chart with party name on X axis and seats won on y axis. Show results of both the states on the same bar plot. Give proper legends  

Madhya Pradesh  
BJP - Win (163) INC - Win (66) BSP – Win ( 0) Others – Win (1)  
Rajasthan  
INC - Win (69) BJP- Win (115) BSP- Win (2) Others-Win (13)


```python
import matplotlib.pyplot as plt

parties = ['BJP', 'INC', 'BSP', 'Others']
mp_seats = [163, 66, 0, 1]
rajasthan_seats = [69, 115, 2, 13]
plt.pie(mp_seats, labels=parties, autopct="%1.2f%%", explode=[0.1 if seat == max(mp_seats) else 0 for seat in mp_seats], colors=["#ff7b00", "blue", "pink", "gray"])
plt.legend()
plt.title("Madhya Pradesh")
plt.show()

plt.pie(rajasthan_seats, labels=parties, autopct="%1.2f%%", explode=[0.1 if seat == max(rajasthan_seats) else 0 for seat in rajasthan_seats], colors=["#ff7b00", "blue", "pink", "gray"])
plt.legend()
plt.title("Rajasthan")
plt.show()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8, 4), dpi=144)

ax = fig.add_subplot(121)
ax.pie(mp_seats, labels=parties, autopct="%1.2f%%", explode=[0.1 if seat == max(mp_seats) else 0 for seat in mp_seats], colors=["#ff7b00", "blue", "pink", "gray"])
ax.legend()
ax.set_title("Madhya Pradesh")

ax=fig.add_subplot(122)
ax.pie(rajasthan_seats, labels=parties, autopct="%1.2f%%", explode=[0.1 if seat == max(rajasthan_seats) else 0 for seat in rajasthan_seats], colors=["#ff7b00", "blue", "pink", "gray"])
ax.legend()
ax.set_title("Rajasthan")
```

```python
import matplotlib.pyplot as plt

plt.bar(parties, mp_seats, width=0.4)
plt.bar([0.4, 1.4, 2.4, 3.4], rajasthan_seats, width=0.4)
plt.legend(["Madhya Pradesh", "Rajasthan"])
plt.show()
```

Q2. Write a python script to
 - Form a list of K random numbers within a limit N where K and N are set by the user. Minimum value of K should be 10.
 - Define a function (or two functions) to return the composite numbers and prime numbers of this list as two separate lists.
 - Determine the square of all prime numbers and square root of all composite numbers
 - Plot both prime numbers Vs their squares and composites Vs their square roots on the same figure object as scatterplots. ( two different subplots on the same figure object)


```python
import random
import math
import matplotlib.pyplot as plt

def isPrime(num: int) -> bool:
    if(num == 0 or num == 1):
        return False
    for i in range(2, num):
        if(num % i == 0):
            return False
    return True

k = int(input("Enter K: "))
n = int(input("Enter N: "))
if(k < 10):
    print("Minimum value of k should be 10")
    exit()

numlist = [math.floor(random.random() * n) + 1 for i in range(k)]

primes = list(filter(isPrime, numlist))
composites = list(filter(lambda n: not isPrime(n) and n != 0 and n != 1, numlist))

primesquares = list(map(lambda n: n ** 2, primes))
compositeroots = list(map(lambda n: n ** 0.5, composites))

fig = plt.figure(figsize=(8, 4), dpi=144)

ax=fig.add_subplot(121)
ax.set_title("Primes vs Squares")
ax.scatter(primes, primesquares)

ax=fig.add_subplot(122)
ax.set_title("Composites vs Square Roots")
ax.scatter(composites, compositeroots)
```
